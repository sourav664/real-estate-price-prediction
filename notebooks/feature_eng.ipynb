```python
import pandas as pd

# Define facility weights
facility_weights = {
    'private pool': 12, 'smart home': 12, 'concierge services': 12, 'infinity swimming pool': 12, 'helipad': 12,
    'wine cellar': 12, 'sky villa': 11, 'skyline view': 11, 'golf course': 11, 'private jaccuzi': 11, 
    'fingerprint access': 10, 'bar/lounge': 10, 'premium branded fittings': 10, 'high ceiling': 10,
    'swimming pool': 9, 'gymnasium': 9, 'club house': 9, 'multipurpose hall': 9, 'private terrace/garden': 9,
    'vaastu compliant': 9, 'large open space': 9, 'library and business centre': 9, 'security (3-tier)': 9,
    'sea facing': 9, 'modular kitchen': 8, 'jogging and strolling track': 8, 'kids play area': 8, 
    'meditation area': 8, 'skydeck': 8, 'piped gas': 8, 'reserved parking': 8, 'lawn with pathway': 7,
    'water front': 7, 'grand entrance lobby': 7, 'cctv camera': 6, 'intercom facility': 6, 'power backup': 6,
    'visitor parking': 6, 'rain water harvesting': 6, 'kids splash pool': 6, 'indoor games room': 5, 
    'badminton court': 5, 'cycling & jogging track': 5, 'waste disposal': 5, 'maintenance staff': 5, 
    'shopping center': 5, 'community hall': 5, 'event space & amphitheatre': 5, 'study room': 5, 'lift': 4,
    'park': 4, 'cafeteria/food court': 4, 'sewage treatment plant': 4, 'water storage': 4, 
    'wheelchair accessibility': 4, 'wastewater recycling': 4
}



# Function to calculate luxury score
def calculate_luxury_score(property_data):
    """
    Calculate the luxury score based on available facilities.

    
    Returns:
    float: Normalized luxury score (0 to 100).
    
    """
    property_data = list(set(property_data))
    actual_score = sum(facility_weights.get(weight, 0) for weight in property_data)
    max_possible_score = sum(facility_weights.values())
    
    # Normalize score to 100-point scale
    luxury_score = (actual_score / max_possible_score) * 100
    return round(luxury_score, 2)

# Example property data (Replace this with actual data from your dataset)
property_example = {
    'Swimming Pool': 1, 'Gymnasium': 1, 'Club House': 1, 'Private Pool': 0, 'Smart Home': 1,
    'Vaastu Compliant': 1, 'Security (3-Tier)': 1, 'Jogging and Strolling Track': 1, 'Private jaccuzi': 0
}

# Calculate luxury score for this example
luxury_score = calculate_luxury_score(property_example)
print("Luxury Score:", luxury_score)

```

```python
import pandas as pd
df = pd.read_csv('real_estatesv6.csv')
```

```python
load_factors  = {
   'bangalore': 0.25,
   'beyond thane': 0.39,
   'mumbai': 0.40,
   'navi mumbai': 0.40,
   "thane": 0.39,
   "pune": 0.35,
   "greater noida": 0.32,
   "gurgaon": 0.28,
   "new delhi": 0.30,
   "noida": 0.32
}
```

```python
temp_df = df[~df['super_built_up_area'].isnull()]
```

```python
value = temp_df.apply(lambda x: round(x['super_built_up_area']/(1 + load_factors[x['regions']])), axis=1)
```

```python
temp_df.head(2)
```

```python
temp_df.loc[temp_df.index, 'carpet_area'] = value
```

```python
price_per_sqft = round((temp_df['price'] * 10000000 ) / temp_df['carpet_area'])
```

```python
temp_df.loc[price_per_sqft.index, 'price_per_sqft'] = price_per_sqft
```

```python
df.loc[temp_df.index, 'carpet_area'] = temp_df['carpet_area']
df.loc[temp_df.index, 'price_per_sqft'] = temp_df['price_per_sqft']
```

```python
df.to_csv('real_estatesv7.csv', index=False)
```
